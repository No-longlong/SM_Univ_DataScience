```python
import pandas as pd
german = pd.read_csv("german_credit.csv")
list(german.columns.values) # 변수명만 확인
german.describe() # 기술통계량 확인
```

```python
import matplotlib.pyplot as plt
```

```python
german.head()
```

```python
# 가장 먼저 쓰는 열 이름이 마지막에 결과로 나온다.
german_grouped = german['Credit Amount'].groupby(german['Type of apartment']) # 아파트 타입별 그룹핑
german_grouped.mean() # 아파트 타입 그룹별 평균
```

```python
# 가장 먼저 쓰는 열 이름이 마지막에 결과로 나온다.
german_grouped = german[['Credit Amount', 'Type of apartment']].groupby(german['Type of apartment']) # 아파트 타입별 그룹핑
german_grouped.mean() # 아파트 타입 그룹별 평균
```

```python
german_grouped2 = german['Credit Amount'].groupby([german['Purpose'],german['Type of apartment']]) # 분산분석을 위해 사용될 그룹
```

```python
bank = pd.read_excel("bank-full.xlsx")
bank
```

```python
bank.head()
bank.info()
bank.describe() # 명목형 변수 제외하고 연속형 변수만 출력
bank.describe(include = 'all') # 명목형까지 포함해서 출력
# unique는 중복 제외하고 카운트한 항목들 
```

```python
bank['marital'].value_counts()
```

```python
f, ax = plt.subplots(1, 2, figsize=(18,8))

bank['y'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('y: yes or no')

bank[['marital', 'y']].groupby('y', as_index=True).count().plot.bar(ax=ax[1])
```
