### 상관계수와 p-value를 데이터프레임 형식으로 만들기

```python
# 피어슨 상관계수 추출을 위하여 결측값 제거
df.dropna(inplace=True)
```

```python
# 하나의 컬럼에 들어갈 예시
corr = stats.pearsonr(df.스프레드, df.cds스프레드)
corr
```

```python
# 기존 데이터의 컬럼 가져오기
column_list = df.columns
```

```python
# 빈 데이터프레임에 컬럼별 스프레드와의 상관관계, p-value 추가
DF = pd.DataFrame()
for i in column_list:
    corr = stats.pearsonr(df.스프레드, df[i])
    DF[i] = corr
```

```python
# 전치
corr_df = np.transpose(DF)
```

```python
# 컬럼명 수정
corr_df.columns = ['상관계수', 'p-value']
```

```python
# 상관계수기준 정렬
corr_df.sort_values(by='상관계수', ascending=False, inplace=True)
```

```python
# 파일 생성
corr_df.to_excel('상관계수와p값.xlsx')
```
