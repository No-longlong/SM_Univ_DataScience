```python
def preprocessing(df):
    
    #df = df.iloc[:,[0, -1]]
    df.columns=['날짜','S&P','항셍','SToxx','코스피','코스닥']
    df['날짜'] = df['날짜'].apply(pd.to_datetime)
    df['월'] = df['날짜'].apply(lambda x : x.year)
    df['달'] = df['날짜'].apply(lambda x : x.month)
    df['월'] = df['월'].astype(str)
    df['달'] = df['달'].astype(str)
    df.drop('날짜', axis=1, inplace=True)
    df['날짜'] = df[['월', '달']].apply(lambda x: '/'.join(x), axis=1)
    df.drop(['월', '달'], axis=1, inplace=True)
    df.set_index('날짜', inplace=True)
    return df
```
