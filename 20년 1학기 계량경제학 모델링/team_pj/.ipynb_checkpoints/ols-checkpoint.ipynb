M2 log - 굉장히 강한 추세이기 때문에 'M2 증가율'로 반드시 써야 한다.

시계열데이터는 그림으로 꼭 봐야 한다. 강한 선형추세는 절대 쓰면 안된다.

M2를 왜 변수들에 다 곱한것인지? M2가 신용스프레드를 축소시킨다고 생각했다. 단기유동성이 풍부하면 신용스프레드가 축소될것이라 생각.

CROSS TERM이 너무 많으면 해석이 너무 어렵다. 대표적인것을 골라라.

M2가 단기유동성으로 보지 말라. M2는 전체통화량의 대표적인 지표일뿐.

cds프리미엄, 외환시장압력지수같은 것들은 스프레드에 영향을 미치는 요인들이다. 이것들은 그 자체로 설명변수에 들어가야 될 것. + 크로스텀도 괜찮다.

기본적인 식을 하나 만들어라. 신용스프레드, vix, 외환시장 압력지수같은것들을 기본모형으로 넣어. 

단기유동성지수변수를 다시 생각해봐라. + 단기유동성의 크로스텀. 

cds프리미엄의 값에 따라서 단기유동성이 신용스프레드에 미치는 영향이 달라질 것 같다.하면 그것을 크로스텀으로 넣으면 되고...

변수들의 선택들이 비슷비슷하다. 좀 구리다. 이런것들을 다 넣는 방법도 있지만,, 이 중에서 평가를해서 설명력이 좋은것들 한두개로 추리는 것도 좋을 것 같다.

식 하나당 크로스텀 하나 정도가 나을 듯. 모든 변수를 크로스텀을 다 해버리면 해석이 안된다.

그럼 적절한 크로스텀 갯수는 몇개? --> 우리들이 효과를 생각해서 판단해야된다. 가장 유동성이 시중에 미치는 영향이 가장 큰게 무엇일지 생각해봐라.

산업생산지수도 필요 없을 것 같다. 굳이 넣은다면 변동률로 바꿔야될것 같다. 

계절조정과 추세는 별개이다. 계절성도 있고 추세도 있기 때문이다. 추세, 계절조정 모두 제거되어야 하는 것이다.

신용스프레드 영향주는 기본 식 하나, 우리가 특화시키고 싶은 것(단기유동성 같은)을 크로스텀으로 만든다.

"효과가 비선형적인 효과를 고려해서 크로스텀을 진행해봤다" 이런식으로.

크로스텀을 안넣은 식으로부터 출발해서 한두개씩 조합하면서 우리가 평가해야 되는 것.

```python
def processSubset(X,y, feature_set):
            model = sm.OLS(y,X[list(feature_set)]) # Modeling
            regr = model.fit() # 모델 학습
            AIC = regr.aic # 모델의 AIC
            return {"model":regr, "AIC":AIC}
```

```python
import time
import itertools
# getBest: 가장 낮은 AIC를 가지는 모델 선택 및 저장
def getBest(X,y,k):
    tic = time.time() # 시작시간
    results = [] # 결과 저장공간
    for combo in itertools.combinations(X.columns.difference(['const']), k): # 각 변수조합을 고려한 경우의 수
        combo=(list(combo)+['const'])
        
        results.append(processSubset(X,y,feature_set=combo))  # 모델링된 것들을 저장
    models = pd.DataFrame(results) # 데이터 프레임으로 변환
    # 가장 낮은 AIC를 가지는 모델 선택 및 저장
    best_model = models.loc[models['AIC'].argmin()] # index
    toc = time.time() # 종료시간
    print("Processed ", models.shape[0], "models on", k, "predictors in", (toc - tic),
          "seconds.")
    return best_model
```

```python
def forward(X, y, predictors):
    # 데이터 변수들이 미리정의된 predictors에 있는지 없는지 확인 및 분류
    remaining_predictors = [p for p in X.columns.difference(['const']) if p not in predictors]
    tic = time.time()
    results = []
    for p in remaining_predictors:
        results.append(processSubset(X=X, y= y, feature_set=predictors+[p]+['const']))
    # 데이터프레임으로 변환
    models = pd.DataFrame(results)

    # AIC가 가장 낮은 것을 선택
    best_model = models.loc[models['AIC'].argmin()] # index
    toc = time.time()
    print("Processed ", models.shape[0], "models on", len(predictors)+1, "predictors in", (toc-tic))
    print('Selected predictors:',best_model['model'].model.exog_names,' AIC:',best_model[0] )
    return best_model
```
