```python
equation_mlb = 'lsalary~games+hruns+rbis+bavg+catcher+hispan'
results = ols(equation_mlb, mlb1).fit()
results.summary()
```

```python
equation_mlb_r = 'lsalary~games+bavg+catcher+hispan'
results_r = ols(equation_mlb_r, mlb1).fit()
results_r.summary()
```
